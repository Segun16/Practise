```python
housing_dummies= pd.get_dummies(housing_c['ocean_proximity'])
```

```python
housing = pd.concat([housing_c, housing_dummies], axis = 1)
```

```python
housing.head()
```

```python
housing.drop(['ocean_proximity'], axis = 1, inplace = True)
```

```python
housing.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = housing.drop(['median_house_value'], axis =1)
y = housing['median_house_value']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lm.fit(X_train,y_train)
```

```python
Prediction = lm.predict(X_test)
```

```python
sns.distplot((y_test-Prediction))
```

```python
print(lm.score(X_test,y_test))
```

```python
sns.regplot(y_test, Prediction)
```

```python
from sklearn import metrics
```

```python
print('RMSE:', np.sqrt( metrics.mean_squared_error(y_test,Prediction)))
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()
```

```python
forest_reg.fit(X_train,y_train)
```

```python
rf_predict = forest_reg.predict(X_test)
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rf_predict)))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parm_grid = [{'n_estimators':[3,8,10], 'max_features':[2,4,6,8]}, {'bootstrap':[False], 'n_estimators':[3,10], 'max_features':[2,3,4]}]
```

```python
grid_search = GridSearchCV(forest_reg, parm_grid, cv=5, scoring = 'neg_mean_squared_error', return_train_score = True)
```

```python
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_result = grid_search.cv_results_
```

```python
for mean_score, params in zip(grid_result['mean_test_score'], grid_result['params']):
    print(np.sqrt(-mean_score), params)
```
